# **Customer Personality Analysis**

### **Business Objectives**
  Customer personality analysis helps a business to modify its product based on its target customers from different types of customer segments.
  
  Finding the potential customers by analysing the behaviour of them is useful to understand the targeted customers.    

  For example, instead of spending money to market a new product to every customer in the company's database, a company can analyse which customer segment is most likely to buy the product and then market the product only on that particular segment.

### **Outlier Detection**

```python
# Find the outliers in the dataset

for col in data.select_dtypes(include=['int64', 'float64']):
  # Calculate the interquartile range (IQR)
  Q1 = data[col].quantile(0.25)
  Q3 = data[col].quantile(0.75)
  IQR = Q3 - Q1

  # Define the lower and upper bounds
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  # Find the outliers
  outliers = data[~((data[col] >= lower_bound) & (data[col] <= upper_bound))]

  # Print the number of outliers for each column
  print(f"Number of outliers for {col}: {len(outliers)}")
```

```python
# Create an empty DataFrame to store data without outliers
data_no_outliers = pd.DataFrame()

for col in data.select_dtypes(include=['int64', 'float64']):
    # Calculate the interquartile range (IQR)
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1

    # Define the lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Remove outliers and append non-outliers to the new DataFrame
    non_outliers = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
    data_no_outliers = pd.concat([data_no_outliers, non_outliers], axis=0)

# Reset index of the new DataFrame
data_no_outliers.reset_index(drop=True, inplace=True)
# Print the shape of the new DataFrame without outliers
print("Shape of data without outliers:", data_no_outliers.shape)
```

# Standardization and Label Encoding

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder

# dividing the data into cont and discrete data using drop function
cont = data.drop(data.columns[[0,1]],axis = 1)
dis = data[data.columns[[0,1]]]
ss = StandardScaler()
standata = ss.fit_transform(cont)
standata
standata = pd.DataFrame(standata)

column_names = cont.columns.tolist()

# Assign column names to df2
standata.columns = column_names
LE = LabelEncoder()
dis.iloc[:,0] = LE.fit_transform(dis.iloc[:,0])
dis.iloc[:,1]= LE.fit_transform(dis.iloc[:,1])
```

Final Transformed Data

```python
# getting full data set after data transformation
customer_final = pd.concat([dis,standata],axis = 1)
customer_final
```

Taking Income and Total Purchases to plot the dendrogams

```python
# construction of Dendogram
X = customer_final[["Income","Total_Purchases"]]
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(X, method='complete'))
```

# AgglomerativeClustering and silhouette_score

```python
## Forming a group using clusters
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')

Y_agg = cluster.fit_predict(X)
```

```python
Y_agg = pd.DataFrame(Y_agg)
Y_agg = Y_agg.rename(columns={0: 'Aggclustering'})
Y_agg['Aggclustering'].value_counts()
```

```python
# Compute the silhouette score
silhouette_avg_agg = silhouette_score(X, Y_agg)
print("Agglomerative Silhouette Score:", silhouette_avg_agg)
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(X.iloc[:,0], X.iloc[:,1], c=cluster.labels_, cmap='rainbow')
```

```python
%%writefile app.py
import numpy as np
import pickle
import streamlit as st
import os

def predict_cluster(Education, Marital_Status, Income, Recency, Complain, Age, Total_Expenses, Total_Purchases, Total_Accepted_Camp, Total_Kids):
    # Map education and marital status to numerical values
    education_mapping = {'Under Graduate': 0, 'Post Graduate': 1}
    marital_status_mapping = {'Single': 0, 'Couple': 1}

    # Convert to numerical values
    Education = education_mapping.get(Education, -1)
    Marital_Status = marital_status_mapping.get(Marital_Status, -1)

    # Check if mappings exist
    if Education == -1 or Marital_Status == -1:
        return "Invalid input"

    # Extract relevant features
    input_data = np.array([float(Income), float(Recency)]).reshape(1, -1)

    # Get the directory of the current script
    script_directory = os.path.dirname(os.path.abspath(__file__))

    # Load KMeans model
    kmeans_model_path = os.path.join(script_directory, "a_kmeans.pkl")

    with open(kmeans_model_path, "rb") as f:
     kmeans_model = pickle.load(f)

    # Predict cluster
    cluster_prediction = kmeans_model.predict(input_data)

    return f'Cluster {cluster_prediction[0]}'

def main():
    st.title("Customer Segmentation")
    st.sidebar.title("Model Parameters")

    # Dropdowns for education and marital status
    Education = st.sidebar.selectbox("Education", ['Under Graduate', 'Post Graduate'])
    Marital_Status = st.sidebar.selectbox("Marital Status", ['Single', 'Couple'])

    # Input fields for other features
    Income = st.sidebar.text_input("Income", "")
    Recency = st.sidebar.text_input("Recency", "")
    Complain = st.sidebar.text_input("Complain", "")
    Age = st.sidebar.text_input("Age", "")
    Total_Expenses = st.sidebar.text_input("Total Expenses", "")
    Total_Purchases = st.sidebar.text_input("Total Purchases", "")
    Total_Accepted_Camp = st.sidebar.text_input("Total Accepted Camp", "")
    Total_Kids = st.sidebar.text_input("Total Kids", "")

    # Button to predict
    if st.sidebar.button("Predict"):
        cluster_prediction = predict_cluster(Education, Marital_Status, Income, Recency, Complain, Age, Total_Expenses, Total_Purchases, Total_Accepted_Camp, Total_Kids)
        st.success('The predicted cluster is: {}'.format(cluster_prediction))

if __name__ == '__main__':
    main()
```
